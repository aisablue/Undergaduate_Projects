Below, I present the general code for the **Newton–Raphson Method**, as I wrote it in Spyder:


```python
from numpy import e, sin, cos

def f(x):
    ''' 2*x + 3*sin(4x) - exp(x) '''
    return 2*x + 3*sin(4*x) - e**x

def df(x):
    ''' 2 + 12*cos(4x) - exp(x) '''
    return 2 + 12*cos(4*x) - e**x

# I want the user to provide the initial reference point
x_0 = float(input('Initial reference point: '))

# Counter for the number of iterations
i = 0

# I want the user to specify the desired accuracy
aq = float(input('Maximum absolute error (in decimal form):\n'))

while True:
    # Terminate the program if the first derivative becomes zero
    if df(x_0) == 0:
        print('The process cannot continue because the first derivative is zero at x_0 =', x_0,
              'with f(x_0) =', f(x_0))
        break
    else:
        i += 1
        x_0 = x_0 - f(x_0) / df(x_0)
    if abs(f(x_0)) <= aq:
        print('Number of iterations:', i, '\n',
              'The root is ρ =', x_0, '\n',
              'With absolute error:', f(x_0), '\n')
        break

print('End of program')
```

### (d)
I will modify the existing code by introducing and checking the relationships between the errors.


I start with the Bisection Method, where I verify the recursive relation in two ways:  
first, by examining the absolute error at each iteration, and second, by observing the number of iterations required to achieve the desired accuracy.


```python
# general code
from numpy import e, sin, log2

def f(x):
    ''' 2*x + 3*sin(4x) - exp(x) '''
    return 2*x + 3*sin(4*x) - e**x

# I want the user to provide the interval endpoints
x_1 = float(input('Lower bound of the interval:\n'))
x_2 = float(input('Upper bound of the interval:\n'))

# ********** I want lists of values
x_list = []
f_list = []

print('f(x_1) =', f(x_1), '\n')
print('f(x_2) =', f(x_2), '\n')

# I want a counter for the iterations
i = 0

# I want the user to provide the desired accuracy
aq = float(input('Maximum absolute error (in decimal form):\n'))

while True:
    i += 1
    x_3 = (x_1 + x_2) / 2
    # ********** Add to the lists the x values and the absolute errors, which coincide with |f(x)|
    x_list.append(x_3)
    f_list.append(abs(f(x_3)))

    if abs(f(x_3)) <= aq:
        break
    elif f(x_3) * f(x_1) < 0:
        x_2 = x_3
    elif f(x_3) * f(x_2) < 0:
        x_1 = x_3

print('Number of iterations:', i, '\n', 'The root is ρ =', x_3, '\n', 'with f(x_3):', f(x_3), '\n')

# ********** Create the recursive-error relation using the list
# ε_{n} = ε_0 / (2^n)
sf_list = [abs(f(x_list[0])) / (2**x) for x in range(0, i)]

print('What we obtained:\n', f_list, '\n')
print('What is predicted:\n', sf_list, '\n')

if f_list == sf_list:
    print('The error relation is confirmed.')
else:
    print('The recursive error relation is not exactly confirmed.')

# Let’s find the theoretical number of iterations:
n = round(log2(abs(f_list[0]) / aq))
print('From theory, for accuracy', aq, 'the required number of iterations is', n)

if n == i:
    print('“Experimental” and theoretical iteration counts are the same.')
else:
    print('The number of iterations in theory and in practice differ.')

print('End of program')
```

\mathbf{x} = \mathbf{g(x)}, \quad \mathbf{x} =
\begin{pmatrix} x \\ y \end{pmatrix}.
$

A natural way to construct such a mapping is to rewrite the system as:

\begin{aligned}
x &= \frac{y + e^{x} - 3}{2}, \\
y &= 0.5x - 2x^{3}.
\end{aligned}

This leads to the fixed–point iteration:

\begin{cases}
x_{n+1} = g_1(x_n, y_n) = \dfrac{y_n + e^{x_n} - 3}{2}, \\
y_{n+1} = g_2(x_n, y_n) = 0.5\,x_n - 2x_n^{3}.
\end{cases}


To investigate convergence, we look at the Jacobian of \(\mathbf{g}\):
$
J_g(x,y) =
\begin{pmatrix}
\frac{\partial g_1}{\partial x} & \frac{\partial g_1}{\partial y} \\
\frac{\partial g_2}{\partial x} & \frac{\partial g_2}{\partial y}
\end{pmatrix}
=
\begin{pmatrix}
\frac{e^{x}}{2} & \frac{1}{2} \\
0.5 - 6x^{2} & 0
\end{pmatrix}.
$

For a fixed–point method to be convergent in a neighbourhood of a solution
$(x^{*}, y^{*})$, we need the mapping to be a contraction there, e.g.
$
\rho\big(J_g(x^{*},y^{*})\big) < 1
$
(where $\rho$ is the spectral radius), or at least some matrix norm
$|J_g(x^{*},y^{*})\| < 1$.

However:

- For the **large positive root** (the one you found with Newton–Raphson, $x^{*} > 6$),
  we have $e^{x^{*}}$ very large, so
 $
  \left|\frac{\partial g_1}{\partial x}(x^{*},y^{*})\right|
  = \frac{e^{x^{*}}}{2} \gg 1.
  $
  This already implies that $J_g$ is not a contraction near that solution,
  so a simple fixed–point iteration based on this \(g\) will **not** converge
  (or will be extremely unstable) for the large positive root.

- For the **negative root** (around $x^{*} \approx -1$), the exponential term is small
  ($e^{x^{*}}/2 < 1$), so it is *possible* to choose a rearrangement $g$
  that satisfies $|\partial g_1 / \partial x| < 1$ and thus converges locally.
  However, this requires a careful choice of the fixed–point form and a suitable
  region where all partial derivatives remain small.

**Conclusion.**  
Although it is in principle possible to construct a fixed–point iteration
$\mathbf{x} = \mathbf{g(x)}$ that converges to **one** of the roots (e.g. the negative one),
for the **positive root** that we computed in part (a) the natural rearrangements lead to
$|\partial g_1 / \partial x| \gg 1$ near the solution, so the fixed–point method is
*not* a contraction and does **not** provide a reliable or efficient way to compute
that solution.  

For this system, the Newton–Raphson method for systems is therefore much more appropriate
and robust than a simple fixed–point iteration.
